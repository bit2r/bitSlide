### 개회사

::: callout-note
#### 개회사 잘 작성하는 방법을 알려줘

개회사를 잘 작성하는 방법은 다음과 같습니다:

1.  목표 및 의도 명확히하기: 개회사의 목적과 이벤트 또는 회의의 주제를 명확하게 전달해야 합니다. 이를 통해 참석자들이 이벤트의 취지를 이해할 수 있습니다.

2.  인사말: 개회사 시작 시 친절하고 정중한 인사말을 통해 참석자들을 환영하고 편안하게 만들어 주세요.

3.  참석자에 대한 감사 표현: 이벤트에 참석한 모든 사람들에게 감사의 인사를 전달하세요. 이는 참석자들에게 소중한 시간을 할애해주었다는 점을 인정해주는 것입니다.

4.  이벤트 개요 제공: 이벤트의 주요 내용, 일정, 참여자 등을 간략하게 소개하세요. 이를 통해 참석자들이 이벤트의 전체적인 구성을 파악할 수 있습니다.

5.  주요 연사 및 게스트 소개: 이벤트에 참석하는 연사나 게스트의 소개를 통해 그들의 전문성과 이벤트와의 연관성을 강조하세요.

6.  중요한 정보 공유: 참석자들이 알아야 할 중요한 사항이나 규정, 안전 사항 등을 안내하세요.

7.  자극적이고 긍정적인 메시지 전달: 개회사를 통해 참석자들에게 동기부여를 주고 이벤트에 대한 기대감을 높여주세요. 이를 위해 긍정적인 메시지와 함께 이벤트의 성공적인 진행을 기대하는 문구를 사용할 수 있습니다.

8.  개회사 마무리: 마지막으로 참석자들에게 좋은 시간을 보내길 바라며, 개회사를 정중하게 마무리하세요.

개회사 작성 시에는 명확하고 간결한 표현을 사용하고, 참석자들의 관심을 끌 수 있는 내용을 담아주세요.
:::

### 지시명령어

::: callout-note
#### 좋은 개회사를 위한 채팅GPT 프롬프트 작성 방법을 알려주세요.

좋은 개회사를 위한 ChatGPT 프롬프트를 작성하려면 문맥을 설정하고, 원하는 내용을 명확하게 지정하며, 명확하고 간결한 언어를 사용해야 합니다. 다음은 따라야 할 단계입니다:

1.  맥락을 설정합니다: 이벤트 유형, 주제 또는 상황 등 필요한 배경 정보를 제공하여 ChatGPT가 요청의 맥락을 이해하는 데 도움이 되도록 합니다.

2.  원하는 콘텐츠를 지정합니다: 시작 문장에 AI가 포함하기를 원하는 측면이나 요소를 명확하게 언급하세요. 여기에는 이벤트의 주제나 목적, 특정 목표 또는 의도된 청중이 포함될 수 있습니다.

3.  명확하고 간결한 언어를 사용하세요: ChatGPT가 이해하기 쉬운 간단하고 직관적인 언어를 사용하여 프롬프트를 작성하세요. 모델을 혼란스럽게 할 수 있는 복잡한 단어, 전문 용어 또는 모호한 문구는 사용하지 마세요.

4.  요청의 틀을 잡으세요: 프롬프트에 질문이나 지시문을 사용하여 ChatGPT가 시작 진술을 제공하기를 원한다는 것을 명확히 하세요.

5.  집중력을 유지하세요: 관련 없는 여러 가지를 묻지 말고 하나의 주제나 요청에 집중하도록 프롬프트에 집중하세요.

다음은 좋은 시작 문장을 위한 잘 작성된 ChatGPT 프롬프트의 예입니다:

"여러분이 지속 가능한 농업에 관한 컨퍼런스의 연사라고 가정해 보세요. 지속 가능한 농업 관행의 중요성, 식량 안보와 환경에 미치는 영향, 그리고 이 컨퍼런스가 이러한 문제를 어떻게 해결하고자 하는지를 간략하게 설명하는 매력적인 시작 멘트를 작성해 주세요."

이 프롬프트는 맥락(지속 가능한 농업 컨퍼런스)을 설정하고, 원하는 콘텐츠(지속 가능한 농업 관행의 중요성, 그 영향, 컨퍼런스 목표)를 명시하며, 명확하고 간결한 언어를 사용하여 ChatGPT가 요청을 쉽게 이해할 수 있도록 합니다.
:::

### AI 기념사 (국문)

신사 숙녀 여러분, 귀빈 여러분, 그리고 동료 인공지능 애호가 여러분, 오늘 경기도 인공지능 서밋의 기조연설자로 이 자리에 서게 되어 진심으로 영광입니다. 존경받는 전문가, 연구자, 업계 리더들이 모인 이 자리는 인공지능이 우리의 세계를 형성하고 집단지성을 발전시키는 데 중요한 역할을 하고 있음을 강조합니다.

활기찬 경기도에 모인 우리는 인공지능 기술의 혁신적 힘을 인식하는 것이 필수적입니다. 일상적인 업무 자동화부터 산업 혁신에 이르기까지 AI는 경제 성장을 촉진하고 우리의 일상 생활을 향상시키는 원동력입니다. 또한 이 획기적인 기술은 기후 변화, 의료, 교육 등 인류의 가장 시급한 과제를 해결할 수 있는 잠재력을 가지고 있습니다.

AI의 영향력은 개인의 삶을 넘어 사회 전체를 재편할 수 있는 역량을 갖추고 있습니다. AI의 힘을 활용하면 새로운 기회를 창출하고 혁신적인 솔루션을 개발하며 모두에게 보다 공평하고 지속 가능한 미래를 보장할 수 있습니다.

풍부한 인재 풀과 탄탄한 인프라, 미래지향적인 리더십을 갖춘 경기도는 AI 혁신의 글로벌 진원지가 될 수 있는 독보적인 위치에 있습니다. 학계, 산업계, 정부 간의 협력을 촉진함으로써 획기적인 연구를 육성하고 최첨단 AI 애플리케이션 개발을 촉진하는 환경을 조성할 수 있습니다.

이 흥미진진한 여정을 함께 시작하면서 AI의 진정한 잠재력은 알고리즘과 기술뿐만 아니라 더 큰 선을 위해 그 힘을 활용할 수 있는 우리의 능력에 있다는 점을 기억합시다. 인공지능이 긍정적인 변화의 촉매제 역할을 하도록 하는 것은 우리 모두의 공동 책임이며, 경기도가 인공지능의 미래를 만들어가는 데 중추적인 역할을 할 것이라 확신합니다.

관심을 가져주셔서 감사드리며, 영감 있고 생산적인 서밋이 되기를 기대합니다.

### 크롤링: 코드

```python
#| eval: false
#| echo: true
#| code-line-numbers: true


import requests
import pandas as pd

# URL of Samsung Electronics stock historical data
url = "https://finance.naver.com/item/sise_day.naver?code=005930&page="

# Empty list to store stock price data
data = []

# Loop through the last 3 years of pages (60 pages in total)
for page in range(1, 61):
    # Send a GET request to the URL with the current page number
    response = requests.get(url + str(page), headers={'User-agent': 'Mozilla/5.0'})
    

    # Check if the response is successful
    if response.status_code == 200:
        # Read the HTML table into a pandas dataframe
        df_list = pd.read_html(response.text, flavor='html5lib')

        # Check if the list of dataframes is not empty
        if df_list:
            # Get the first dataframe
            df = df_list[0]

            # Drop the last row which is just a duplicate of the column headers
            df = df.drop(len(df) - 1)

            # Append the dataframe to the list
            data.append(df)

# Concatenate all dataframes into a single dataframe
df = pd.concat(data)

# Reverse the order of rows to start from the earliest date
df = df.iloc[::-1]

df = df.dropna()

# Set the date column as the index and remove unnecessary columns
df = df.set_index('날짜')[['종가', '거래량']]

# Convert the data types of columns to float and int
df['종가'] = df['종가'].astype(float)
df['거래량'] = df['거래량'].astype(int)

# Save the dataframe as a CSV file
df.to_csv('data/samsung.csv')

# Print a confirmation message
print('samsung.csv saved successfully.')
```

### 문제

#### 디오판토스 묘비

신의 축복으로 태어난 그는 인생의 $\frac{1}{6}$을 소년으로 보냈다. 그리고 다시 인생의 $\frac{1}{12}$ 이 지난 뒤에는 얼굴에 수염이 자라기 시작했다. 다시 $\frac{1}{7}$이 지난 뒤 그는 아름다운 여인을 맞이하여 화촉을 밝혔으며, 결혼한 지 5년 만에 귀한 아들을 얻었다. 아! 그러나 그의 가엾은 아들은 아버지의 반 밖에 살지 못했다. 아들을 먼저 보내고 깊은 슬픔에 빠진 그는 그 뒤 4년간 정수론에 몰입하여 스스로를 달래다가 일생을 마쳤다. [^1]

### 수식 표현

디오판토스가 정확히 언제 태어나고 언제 죽었는지는 명확하지 않지만, 그가 죽었을 때의 나이는 정확히 알 수 있다. 이를 방정식을 통해 풀어보자. 수식을 $x$에 대해 정리해서 풀면 84가 나온다.

$$\frac {x} {6} + \frac {x} {12} + \frac {x} {7} +  5 + \frac {x} {2} + 4 = x$$

### 손으로 풀기

![](../../img/inference_hand.jpg){width="486"}

### `Sympy` 동원

```python
#| echo: true
from sympy import *
x = Symbol('x')
Diophantus_puzzle = x/6 + x/12 + x/7 + 5 + x/2 + 4 - x
solve(Diophantus_puzzle, x, dict = True)
```
